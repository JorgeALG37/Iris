Observemos que en la longitud del sÃ©palo se tiene un valor de la especie **Iris virginica** que se asemeja mÃ¡s a los valores de la especie **Iris setosa**, por lo que podrÃ­a sesgar nuestra predicciÃ³n. \\
AdemÃ¡s, en el ancho del sÃ©palo tenemos un valor que se aleja de manera considerable a los valores de la especie **Iris setosa**, por lo que tambiÃ©n podrÃ­a sesgar nuestra predicciÃ³n.

```python
print("Data Frame antes de la eliminaciÃ³n de valores: " + str(data.shape))
data = data[(data['Species']!='Iris-virginica') | (data['SepalLengthCm']>=5)]
data = data[(data['Species']!='Iris-setosa') | (data['SepalWidthCm']>=2.5)]
print("Data Frame despuÃ©s de la eliminaciÃ³n de valores: " + str(data.shape))
```

#Machine Learning

```python
from sklearn.preprocessing import LabelEncoder

# Crear un LabelEncoder
le = LabelEncoder()

# Adapta y transforma las variables categÃ³ricas
data['Species'] = le.fit_transform(data['Species'])
```

```python
sns.heatmap(data.corr(), annot=True, cmap='Blues')
plt.show()
```

Notemos que la variable que tiene mayor correlaciÃ³n son el ancho de pÃ©talo, despuÃ©s longitud de pÃ©talo, y le sigue la longitud de sÃ©palo. Por otro lado, aquella variable que no determina la especie de la flor es el ancho de sÃ©palo.

##CreaciÃ³n del modelo

```python
#Determinar las variables predictoras y la variable a predecir
X = data.drop('Species', axis=1)
Y = data['Species']
```

```python
#Configurar los conjuntos de entrenamiento y validaciÃ³n
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
#Estandarizar los datos
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

#Inicializar el modelo
knn = KNeighborsClassifier(n_neighbors=5)

#Entrenar el modelo
knn.fit(X_train,Y_train)
```

##Evaluar el modelo

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#PredicciÃ³n
y_pred = knn.predict(X_test)
y_pred
```

```python
#Accuracy_score
print("PrecisiÃ³n del modelo:", accuracy_score(Y_test, y_pred))
```

```python
#Classification_report
print("Reporte de clasificaciÃ³n: \n", classification_report(Y_test,y_pred))
```

```python
#Confusion_matrix
mc = confusion_matrix(Y_test, y_pred)

plt.figure(figsize=(10,5))
sns.heatmap(mc, annot=True, cbar=False, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicciones')
plt.ylabel('Valores reales')
plt.show()

#print("Matriz de confusiÃ³n: \n", confusion_matrix(Y_test, y_pred))
```

Podemos observar que el modelo tiene una precisiÃ³n del 100%, ya que predijo todos los valores de manera correcta.
